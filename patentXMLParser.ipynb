```python
#encoding=utf8
from lxml import etree
parser=etree.XMLParser(load_dtd= True)
tree = etree.parse("DOCDB-201511-017-KR-0129.xml",parser)
root = tree.getroot()
ns = {"exch": "http://www.epo.org/exchange"}
```

```python
subroot = root.getchildren()
```

```python
len(subroot)
```

```python
subroot_check = etree.tostring(subroot[0])
print(subroot_check)
```

```python
def els(ipcrs): # 处理IPC数据
    els = []
    for el in ipcrs:
        els.append(el.find('.//text').text)
    return els
def elv(inventors): #处理INV数据
    elv = []
    for el in inventors:
        elv.append(el.find('.//exch:inventor-name/name',ns).text+' ('+el.find('.//residence/country').text+','+el.get('sequence')+')')
    return elv
def ela(applicants): #处理APP数据
    ela = []
    for el in applicants:
        ela.append(el.find('.//exch:applicant-name/name',ns).text+' ('+el.find('.//residence/country').text+','+el.get('sequence')+')')
    return ela
```

```python
from pandas.core.frame import DataFrame
patents = []
for document in subroot:
    country = document.get("country")
    docnumber = document.get("doc-number")
    kind = document.get("kind")
    docid = document.get("doc-id")
    datepubl = document.get("date-publ")
    familyid = document.get("family-id")
    representative = document.get("is-representative")
    lastexchange = document.get("date-of-last-exchange")
    addeddocdb = document.get("date-added-docdb")
    originatingoffice = document.get("originating-office")
    for node in document.findall('.//exch:publication-reference[@data-format="docdb"]/document-id/country',namespaces=ns):
        pubcountry = node.text
    for node in document.findall('.//exch:publication-reference[@data-format="docdb"]/document-id/doc-number',namespaces=ns):
        pubnumber = node.text
    for node in document.findall('.//exch:publication-reference[@data-format="docdb"]/document-id/doc-number',namespaces=ns):
        pubkind =  node.text
    for node in document.findall('.//exch:publication-reference[@data-format="docdb"]/document-id/date',namespaces=ns):
        pubdate = node.text
    for node in document.findall('.//exch:application-reference[@data-format="docdb"]/document-id/country',namespaces=ns):
        appcountry = node.text
    for node in document.findall('.//exch:application-reference[@data-format="docdb"]/document-id/doc-number',namespaces=ns):
        appnumber = node.text
    for node in document.findall('.//exch:application-reference[@data-format="docdb"]/document-id/doc-number',namespaces=ns):
        appkind =  node.text
    for node in document.findall('.//exch:application-reference[@data-format="docdb"]/document-id/date',namespaces=ns):
        appdate = node.text
    for node in document.findall('.//exch:bibliographic-data/exch:invention-title',namespaces=ns):
        title = node.text
    for node in document.findall('.//exch:bibliographic-data/exch:invention-title',namespaces=ns):
        title_lg = node.get("lang")
    for node in document.findall('.//exch:abstract/exch:p',namespaces=ns):
        abstract = node.text
    ipcrs = document.findall('.//classification-ipcr')
    inventors = document.findall('.//exch:inventor[@data-format="docdb"]',ns)
    applicants = document.findall('.//exch:applicant[@data-format="docdb"]',ns)
    patent =  {'country':country,'doc-number':docnumber,'kind':kind,'doc-id':docid,'date-publ':datepubl,'family-id':familyid,
              'is-representative':representative,'date-of-last-exchange':lastexchange,'date-added-docdb':addeddocdb,
              'originating-office':originatingoffice,'pub_country':pubcountry,'pub_number':pubnumber,'pub-kind':pubkind,
               'pub-date':pubdate,'app_country':appcountry,'app_number':appnumber,'pub-kind':appkind,
               'app-date':appdate,'ipcrsa':els(ipcrs),'inventors':elv(inventors),'applicants':ela(applicants),'title':title,
               'title_lg':title_lg,'abstract':abstract}
    patents.append(patent)
patents = DataFrame(patents)
```
